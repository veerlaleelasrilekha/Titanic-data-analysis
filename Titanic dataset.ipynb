```python
import pandas as pd
import numpy as np
df=pd.read_csv("tested.csv")
df.fillna(0,inplace=True)
mean_age = df['Age'].replace(0, np.nan).mean()
df['Age'] = df['Age'].replace(0, mean_age)
df
female=(df['Sex']=='female').sum()
male=(df['Sex']=='male').sum()
print("female:",female)
print("male:",male)
female_or_male=(female)<=(male)
print("Male Survived more:",male)
df
```

```python
import matplotlib.pyplot as plt

female_survived     = ((df['Sex'] == 'female') & (df['Survived'] == 1)).sum()
female_not_survived = ((df['Sex'] == 'female') & (df['Survived'] == 0)).sum()
male_survived       = ((df['Sex'] == 'male')   & (df['Survived'] == 1)).sum()
male_not_survived   = ((df['Sex'] == 'male')   & (df['Survived'] == 0)).sum()

plt.bar([0, 1],        [female_survived,     male_survived],     width=0.35, color='#3266ad', label='Survived')
plt.bar([0.35, 1.35],  [female_not_survived, male_not_survived], width=0.35, color='#d0d0c8', label='Did not survive')

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks([0.175, 1.175], ['Female', 'Male'])
plt.legend()
plt.show()
```
